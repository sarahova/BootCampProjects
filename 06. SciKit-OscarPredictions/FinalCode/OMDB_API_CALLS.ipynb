```python
import requests
import pandas as pd
from config import omdb_key
```

```python
#import csv of oscar winners
oscar_movie_db=pd.read_csv("../DataSources/Oscars1934-2015.csv")
pd.to_numeric(oscar_movie_db.Year, errors='coerce')

#Ensure that all the "Names" in the oscar_movie_db are formated correctly (no spaces at end of name)
for i,row in oscar_movie_db.iterrows():    
    j = oscar_movie_db['Name'].iloc[i]
    if j[-1] != " ":
        pass
    else:
        j=j[:-1]
        oscar_movie_db['Name'].iloc[i]=j

#fix Moulin Rouge 1952 and Moulin Rouge 2001 (remake of movie) so that these movies are treated distinct
for i,film in oscar_movie_db.iterrows():
    if oscar_movie_db['Name'][i]=='Moulin Rouge' and oscar_movie_db['Year'][i]==1952:
        oscar_movie_db['Name'][i]='Moulin Rouge 1952'
    elif oscar_movie_db['Name'][i]=='Moulin Rouge' and oscar_movie_db['Year'][i]==2001:
        oscar_movie_db['Name'][i]='Moulin Rouge 2001'
    elif oscar_movie_db['Film'][i]=='Moulin Rouge' and oscar_movie_db['Year'][i]==1952:
        oscar_movie_db['Film'][i]='Moulin Rouge 1952'
    elif oscar_movie_db['Film'][i]=='Moulin Rouge' and oscar_movie_db['Year'][i]==2001:
        oscar_movie_db['Film'][i]='Moulin Rouge 2001'      

#create a movie list of best picture movie winners and nominees list
df_movies = oscar_movie_db[((oscar_movie_db['Award']== 'Best Picture')|(oscar_movie_db['Award']== 'Best Motion Picture'))]
df_movies=df_movies.reset_index(drop=True)
movies= list(df_movies['Name'])
```

```python
#OMBD Calls - Movies
theMovies_OMBD = []

for movie in movies:
    if movie=='Moulin Rouge 1952':
        baseUrl_OMBD="http://www.omdbapi.com/?t=Moulin+Rouge&y=1952&apikey="+omdb_key    
    elif movie=='Moulin Rouge 2001':
        baseUrl_OMBD="http://www.omdbapi.com/?t=Moulin+Rouge&y=2001&apikey="+omdb_key
    else:
        baseUrl_OMBD = "http://www.omdbapi.com/?t=" + movie + "&apikey="+omdb_key

    try:
        movieData_OMBD = requests.get(baseUrl_OMBD).json()
        title = movieData_OMBD['Title']    
        year = movieData_OMBD['Year']
        rated = movieData_OMBD['Rated']
        genre = movieData_OMBD['Genre']
        director = movieData_OMBD['Director']
        writer = movieData_OMBD['Writer']
        actors = movieData_OMBD['Actors']
        country = movieData_OMBD['Country']
        awards = movieData_OMBD['Awards']
        imdbRating = movieData_OMBD['imdbRating']
        boxOffice = movieData_OMBD['BoxOffice']
        releaseDate=movieData_OMBD['Released']
        
        #movieData
        theMovie = {
        'Title': title,
        'Year': year,
        'Rated': rated,
        'Genre': genre,
        'Director': director,
        'Writer(s)': writer,
        'Actors': actors,
        'Country': country,
        'Awards': awards,
        'imdbRating': imdbRating,
        'Box Office': boxOffice,
        'Release Date': releaseDate
        }
        theMovies_OMBD.append(theMovie)
    
    except:
        pass
    
theMoviesDF_OMBD = pd.DataFrame(theMovies_OMBD)
theMoviesDF_OMBD.to_csv("../DataSources/theMoviesDF_OMBD.csv", index=False, encoding='utf-8')
theMoviesDF_OMBD
```
